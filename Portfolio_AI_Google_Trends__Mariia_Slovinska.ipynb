## üéØ Project Objective

To analyze and visualize the popularity trends of emerging AI tools using Google Trends data for the past year.  
The goal is to identify how interest in different tools evolves over time and to draw insights for product teams, market researchers, and technology enthusiasts.

## üß∞ Tools Used

- **Google Trends**: For extracting interest over time.
- **Pandas**: For data transformation and cleaning.
- **Seaborn & Matplotlib**: For data visualization.
- **Jupyter Notebook**: For project execution and documentation.

---


```python

import pandas as pd
import matplotlib.pyplot as plt

# read combined file
df = pd.read_csv("combined_google_trends.csv", index_col=0, parse_dates=True)
df = df.sort_index()
df.tail()
```

## üîù Top 5 AI Tools by Average Interest

```python

top5 = df.mean().sort_values(ascending=False).head(5).index
df[top5].plot(figsize=(14,6), marker='o', title="Top 5 AI Tools (Avg Interest)")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## üîº Top 5 AI Tools by Growth (First ‚Üí Last Week)

```python

delta = df.iloc[-1] - df.iloc[0]
top_rising = delta.sort_values(ascending=False).head(5).index
df[top_rising].plot(figsize=(14,6), marker='o', title="Top 5 Fastest Growing AI Tools")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## üß± All Tools

```python

cols = df.columns
n = len(cols)
rows = (n // 4) + 1

fig, axes = plt.subplots(rows, 4, figsize=(20, rows * 3), sharex=True)
axes = axes.flatten()

for i, col in enumerate(cols):
    df[col].plot(ax=axes[i], title=col, marker='o')
    axes[i].grid(True)

# Hide plots with null 
for j in range(i+1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

‚úÖ 1. Heatmap displays the level of interest by time and tool

```python
import seaborn as sns

# Create title for X: Display the name of the month only during every 4th week.
xticks = [d.strftime('%b') if i % 4 == 0 else '' for i, d in enumerate(df.index)]

plt.figure(figsize=(14, 8))
ax = sns.heatmap(df.T, cmap="YlGnBu", annot=False, fmt="d")

# –í—Å—Ç–∞–Ω–æ–≤–ª—é—î–º–æ –æ–Ω–æ–≤–ª–µ–Ω—ñ –ø—ñ–¥–ø–∏—Å–∏ –¥–æ –æ—Å—ñ X
ax.set_xticklabels(xticks, rotation=45, ha='right')

plt.title("Google AI Trends Heatmap April 2024 - April 2025")
plt.xlabel("Month")
plt.ylabel(" AI Tool")
plt.tight_layout()
plt.show()
```

‚ûï Noticeable spikes in Google searches in 2025 across all tools.

‚úÖ 2. Bar chart 

```python
df.mean().sort_values().plot(kind='barh', figsize=(12, 8), color='skyblue', title="Avg Interest by AI Tool")
plt.xlabel("Average Interest (last 12 months)")
plt.tight_layout()
plt.show()
```

‚ûï The ranking is clearly visible:

The highest average interest is in Langchain and the AI plugin.

The lowest average interest is in Windsurf AI and the Qodo plugin.
