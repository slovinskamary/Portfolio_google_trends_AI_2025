## üîß Import and Connect Google Trends

```python

from pytrends.request import TrendReq
import pandas as pd
import matplotlib.pyplot as plt
import time

pytrends = TrendReq(hl='en-US', tz=360)
```

## üß† Step 1: Add the first keyword

```python

keyword = "Langchain"  # ‚Üê Change to the next keyword
pytrends.build_payload([keyword], timeframe='today 12-m')
```

## üìä Step 2: Retrieve the time trend and save it if data is available

```python

df = pytrends.interest_over_time()

if not df.empty and keyword in df.columns:
    df = df[[keyword]]
    filename = f"trend_{keyword.replace(' ', '_')}.csv"
    df.to_csv(filename)
    print(f"‚úÖ –î–∞–Ω—ñ –∑–Ω–∞–π–¥–µ–Ω–æ —Ç–∞ –∑–±–µ—Ä–µ–∂–µ–Ω–æ —è–∫ {filename}")
    display(df.head())
else:
    print(f"‚ö†Ô∏è –ù–µ–º–∞—î –¥–∞–Ω–∏—Ö –¥–ª—è –∫–ª—é—á–æ–≤–æ–≥–æ —Å–ª–æ–≤–∞: '{keyword}'. –°–ø—Ä–æ–±—É–π —ñ–Ω—à–µ.")
```

## üìà Step 3: Create plot if data is available

```python

if not df.empty and keyword in df.columns:
    df.plot(figsize=(12,5), title=f"Google Trends: {keyword}")
    plt.ylabel("Interest over time")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("‚ùå –ì—Ä–∞—Ñ—ñ–∫ –Ω–µ –∑–±—É–¥–æ–≤–∞–Ω–æ ‚Äî –Ω–µ–º–∞—î –¥–∞–Ω–∏—Ö.")
```
